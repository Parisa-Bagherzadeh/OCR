```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import easyocr
```

```python
reader = easyocr.Reader(['en'])
img = '/content/images/latin_licenseplate.jpeg'
result = reader.readtext(img)
img = mpimg.imread(img)
plt.figure(figsize=(3, 3))
plt.imshow(img)
plt.axis('off')
plt.title(result[0][1] + ' ' + result[1][1])
plt.show()

```

```python
reader = easyocr.Reader(['en'])
img = '/content/images/latin_handwriting.jpg'
result = reader.readtext(img)
img = mpimg.imread(img)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.axis('off')
title = []

for i in range(len(result)) :
  title.append(result[i][1] )
plt.title(title )
plt.show()

```
